ANALISIS KONSEP: DERET FIBONACCI REKURSIF
------------------------------------------
Berdasarkan materi presentasi, berikut adalah pembedahan konsepnya:

1. DEFINISI REKURSI (Recursion):
   Rekursi adalah teknik di mana sebuah fungsi memanggil dirinya sendiri.
   Dalam kode ini, fungsi 'fibonacci(N)' memanggil 'fibonacci(N-2)'
   dan 'fibonacci(N-1)'.

2. KOMPONEN UTAMA REKURSI:
   A. Base Case (Kondisi Berhenti):
      `if N <= 1: return N`. Tanpa ini, fungsi akan mengalami 'Infinite Recursion'
      (pengulangan tanpa henti) dan menyebabkan error 'RecursionError'.
   B. Recursive Step (Langkah Rekursif):
      `return fibonacci(N-2) + fibonacci(N-1)`. Ini adalah inti dari rumus
      Fibonacci: angka saat ini adalah jumlah dari dua angka sebelumnya.

3. ALUR LOGIKA (Tracing N=4):
   - fibonacci(4) memecah diri menjadi fib(2) + fib(3)
   - fib(3) memecah diri menjadi fib(1) + fib(2)
   - fib(2) memecah diri menjadi fib(0) + fib(1)
   - Setelah mencapai nilai 0 atau 1 (Base Case), nilai-nilai tersebut
     dijumlahkan kembali ke atas (Backtracking).

4. EFISIENSI (Big O Notation):
   Konsep rekursi murni ini memiliki kompleksitas waktu O(2^n).
   Artinya, setiap kenaikan N akan menggandakan waktu komputasi.
   Sangat tidak efisien untuk N > 40.
"""

def fibonacci(N):
    # Base case: Jika N adalah 0 atau 1, kembalikan nilai tersebut
    if N <= 1:
        return N
    else:
        # Recursive step: Memanggil fungsi itu sendiri
        return fibonacci(N-2) + fibonacci(N-1)

# --- BAGIAN EKSEKUSI ---
def main():
    print("=== PROGRAM DERET FIBONACCI REKURSIF ===")
    try:
        # Meminta input N dari pengguna
        n_input = int(input('Masukan nilai N (contoh: 4): '))

        print(f"\nProses Iterasi untuk N = {n_input}:")
        print("-" * 30)

        # Melakukan perulangan untuk mencetak deret seperti pada gambar
        for i in range(1, n_input + 1):
            hasil = fibonacci(i)
            print(f"Iterasi {i} = i = {i} maka hasil: {hasil}")

        print("-" * 30)
        print("Selesai.")

    except ValueError:
        print("Input salah! Harap masukkan angka bulat.")

if __name__ == "__main__":
    main()
