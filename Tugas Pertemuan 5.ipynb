```
# Data mahasiswa
nama = ["Asep", "Budi", "Cecep"]

nilai = [
    [50, 70, 40, 80],  # Asep
    [78, 78, 80, 65],  # Budi
    [57, 88, 67, 69]   # Cecep
]

rata_mahasiswa = []

for i in range(len(nilai)):
    total = sum(nilai[i])
    rata = total / len(nilai[i])
    rata_mahasiswa.append(rata)

maks = max(rata_mahasiswa)
index_mhs = rata_mahasiswa.index(maks)

rata_mk = []

for j in range(len(nilai[0])):  # jumlah kolom (MK)
    total = 0
    for i in range(len(nilai)):
        total += nilai[i][j]
    rata = total / len(nilai)
    rata_mk.append(rata)

min_mk = min(rata_mk)
index_mk = rata_mk.index(min_mk)

print("Mahasiswa Terpintar :", nama[index_mhs], "(Nilai :", round(maks, 2), ")")
print("Mata Kuliah Nilai Terkecil : MK", index_mk + 1, "(Nilai :", round(min_mk, 2), ")")
```

```
# fungsi input matriks
def input_matriks(baris, kolom):
    matriks = []
    for i in range(baris):
        row = []
        for j in range(kolom):
            nilai = int(input(f"Masukkan nilai [{i}][{j}]: "))
            row.append(nilai)
        matriks.append(row)
    return matriks


# fungsi tampil matriks
def tampil(m):
    for row in m:
        print(row)


# penjumlahan
def tambah(A, B):
    hasil = []
    for i in range(len(A)):
        row = []
        for j in range(len(A[0])):
            row.append(A[i][j] + B[i][j])
        hasil.append(row)
    return hasil


# pengurangan
def kurang(A, B):
    hasil = []
    for i in range(len(A)):
        row = []
        for j in range(len(A[0])):
            row.append(A[i][j] - B[i][j])
        hasil.append(row)
    return hasil


# perkalian
def kali(A, B):
    hasil = []
    for i in range(len(A)):
        row = []
        for j in range(len(B[0])):
            total = 0
            for k in range(len(B)):
                total += A[i][k] * B[k][j]
            row.append(total)
        hasil.append(row)
    return hasil



# MENU
while True:
    print("\n=== MENU MATRIKS ===")
    print("1. Penjumlahan")
    print("2. Pengurangan")
    print("3. Perkalian")
    print("0. Exit")

    pilih = input("Pilih menu: ")

    if pilih == "1" or pilih == "2":
        baris = int(input("Masukkan jumlah baris: "))
        kolom = int(input("Masukkan jumlah kolom: "))

        print("\nMatriks A")
        A = input_matriks(baris, kolom)

        print("\nMatriks B")
        B = input_matriks(baris, kolom)

        if pilih == "1":
            hasil = tambah(A, B)
            print("\nHasil Penjumlahan:")
        else:
            hasil = kurang(A, B)
            print("\nHasil Pengurangan:")

        tampil(hasil)

    elif pilih == "3":
        baris_A = int(input("Baris A: "))
        kolom_A = int(input("Kolom A: "))
        baris_B = int(input("Baris B: "))
        kolom_B = int(input("Kolom B: "))

        if kolom_A != baris_B:
            print("Tidak bisa dikalikan! (kolom A harus = baris B)")
        else:
            print("\nMatriks A")
            A = input_matriks(baris_A, kolom_A)

            print("\nMatriks B")
            B = input_matriks(baris_B, kolom_B)

            hasil = kali(A, B)
            print("\nHasil Perkalian:")
            tampil(hasil)

    elif pilih == "0":
        print("Program selesai.")
        break

    else:
        print("Pilihan tidak valid!")
```
